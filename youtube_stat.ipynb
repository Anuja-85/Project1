# Data Preprocessing¶
# Handle Missing Values

```python
# Filling missing values for numerical columns with the mean
mean_values = df.select_dtypes(include=['float64', 'int64']).mean()
df.fillna(mean_values, inplace=True)
```

```python
# Filling missing values for categorical columns with the mode
mode_values = df.select_dtypes(include=['object']).mode().iloc[0]
df.fillna(mode_values, inplace=True)
```

```python
df.isnull().sum()
```

```python
# Remove Duplicates

# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
import plotly.express as px
```

```python
## Implement Filters for Categories, Countries, and Channel Types

# Function to filter data and plot
def plot_filtered_data(category=None, country=None, channel_type=None):
    filtered_df = df

    if category:
        filtered_df = filtered_df[filtered_df['category'] == category]
    if country:
        filtered_df = filtered_df[filtered_df['Country'] == country]
    if channel_type:
        filtered_df = filtered_df[filtered_df['channel_type'] == channel_type]

    fig = px.bar(filtered_df, x='Youtuber', y='subscribers', color='category')
    fig.show()
```

```python
# Example usage:
plot_filtered_data(category='Music', country='United States')
```
